```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import warnings

warnings.filterwarnings('ignore')
```

# Exploratory Data Analysis (EDA)

```python
data = pd.read_csv('final_merged_data.csv')
data['Date'] = pd.to_datetime(data['Date'])
data.info()
```

```python
data.describe()
```

```python
pd.set_option('display.max_columns', None)
# Display describe output
print(data.describe())
```

```python
data.describe().transpose()
```

```python
data.isna().sum()
```

```python
# Plot histograms
data.hist(bins=30, figsize=(15, 10))
plt.tight_layout()
plt.show()
```

```python
# Plot time series
plt.figure(figsize=(15, 15))

# VEDL stock data
plt.subplot(4, 1, 1)
plt.plot(data['Date'], data['Close_VEDL'], label='VEDL Close')
plt.title('VEDL Close Price Over Time')
plt.xlabel('Date')
plt.ylabel('Close Price')
plt.legend()
plt.grid()

# Nifty Metal stock data
plt.subplot(4, 1, 2)
plt.plot(data['Date'], data['Close_NiftyMetal'], label='Nifty Metal Close', color='orange')
plt.title('Nifty Metal Close Price Over Time')
plt.xlabel('Date')
plt.ylabel('Close Price')
plt.legend()
plt.grid()

# Macroeconomic data
plt.subplot(4, 1, 3)
plt.plot(data['Date'], data['GDP_Growth_Rate'], label='GDP Growth Rate', color='green')
plt.plot(data['Date'], data['Inflation_Rate'], label='Inflation Rate', color='red')
plt.title('GDP Growth Rate and Inflation Rate Over Time')
plt.xlabel('Date')
plt.ylabel('Rate')
plt.legend()
plt.grid()

# News Sentiment
plt.subplot(4, 1, 4)
plt.plot(data['Date'], data['News Sentiment'], label='News Sentiments', color='purple')
plt.title('News Sentiment Over Time')
plt.xlabel('Date')
plt.ylabel('News Sentiment')
plt.legend()
plt.grid()

plt.tight_layout()
plt.show()
```

```python
# Scatter plots
plt.figure(figsize=(15, 10))

# VEDL Close vs Nifty Metal Close
plt.subplot(2, 2, 1)
plt.scatter(data['Close_VEDL'], data['Close_NiftyMetal'], alpha=0.5)
plt.title('VEDL Close vs Nifty Metal Close')
plt.xlabel('VEDL Close')
plt.ylabel('Nifty Metal Close')
plt.grid()

# GDP Growth Rate vs Inflation Rate
plt.subplot(2, 2, 2)
plt.scatter(data['GDP_Growth_Rate'], data['Inflation_Rate'], alpha=0.5, color='green')
plt.title('GDP Growth Rate vs Inflation Rate')
plt.xlabel('GDP Growth Rate')
plt.ylabel('Inflation Rate')
plt.grid()

# VEDL Close vs News Sentiment
plt.subplot(2, 2, 3)
plt.scatter(data['Close_VEDL'], data['News Sentiment'], alpha=0.5, color='red')
plt.title('VEDL Close vs News Sentiment')
plt.xlabel('VEDL Close')
plt.ylabel('News Sentiment')
plt.grid()

# Nifty Metal Close vs News Sentiment
plt.subplot(2, 2, 4)
plt.scatter(data['Close_NiftyMetal'], data['News Sentiment'], alpha=0.5, color='purple')
plt.title('Nifty Metal Close vs News Sentiment')
plt.xlabel('Nifty Metal Close')
plt.ylabel('News Sentiment')
plt.grid()

plt.tight_layout()
plt.show()
```

```python
# List of columns to plot
columns_to_plot = ['Open_VEDL', 'High_VEDL', 'Low_VEDL', 'Close_VEDL', 'Adj Close_VEDL', 'Volume_VEDL',
                   'Open_NiftyMetal', 'High_NiftyMetal', 'Low_NiftyMetal', 'Close_NiftyMetal',
                   'Adj Close_NiftyMetal', 'Volume_NiftyMetal', 'GDP_Growth_Rate', 'Inflation_Rate', 'News Sentiment']

# Plotting box plots for each variable
plt.figure(figsize=(20, 30))
for i, column in enumerate(columns_to_plot, 1):
    plt.subplot(8, 2, i)
    plt.boxplot(data[column].dropna())
    plt.title(column)
    plt.grid()

plt.tight_layout()
plt.show()
```

```python
# Compute correlation matrix
corr_matrix = data.corr()

# Plot correlation matrix
plt.figure(figsize=(15, 10))
sns.heatmap(corr_matrix, annot=True, cmap='coolwarm')
plt.title('Correlation Matrix')
plt.show()
```

```python
# Pairplot for selected features
sns.pairplot(data[['Close_VEDL', 'Close_NiftyMetal', 'GDP_Growth_Rate', 'Inflation_Rate', 'News Sentiment']])
plt.suptitle("Pairwise Relationships", y=1.02)
plt.show()
```

```python
# VEDL closing price over time
plt.figure(figsize=(10, 4))
plt.plot(data['Date'], data['Close_VEDL'], label='VEDL Close Price')
plt.title("VEDL Closing Price Trend")
plt.xlabel("Date")
plt.ylabel("Price")
plt.grid(True)
plt.legend()
plt.tight_layout()
plt.show()
```
